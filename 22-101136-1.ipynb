```python
column1 ='How effective do you perceive homemade recipes \nOn a scale from "1-Not Effective" to "5-Highly Effective,"'
column2 ='How effective do you perceive traditional medicines\nOn a scale from "1-Not Effective" to "5-Highly Effective,"\n'
correlation = df[column1].corr(df[column2])
print(correlation)

plt.figure(figsize=(8, 6))  # Adjust figure size as desired
plt.scatter(df[column1], df[column2])  # Plot data points

plt.xlabel(column1)
plt.ylabel(column2)
plt.title(f"Correlation between {column1} and {column2}")

plt.show()
```

```python
print(df.isnull().sum())
```

```python
# Select the columns
homemade_column = df['How effective do you perceive homemade recipes \nOn a scale from "1-Not Effective" to "5-Highly Effective,"']
traditional_column = df['How effective do you perceive traditional medicines\nOn a scale from "1-Not Effective" to "5-Highly Effective,"\n']

# Calculate the mean
homemade_mean = homemade_column.mean()
traditional_mean = traditional_column.mean()
print("Mean of homemade recipes:", homemade_mean)
print("Mean of traditional medicines:", traditional_mean)

# Calculate the median
homemade_median = homemade_column.median()
traditional_median = traditional_column.median()
print("Median of homemade recipes:", homemade_median)
print("Median of traditional medicines:", traditional_median)

# Calculate the mode
homemade_mode = homemade_column.mode()[0]  # Assuming single mode
traditional_mode = traditional_column.mode()[0]  # Assuming single mode
print("Mode of homemade recipes:", homemade_mode)
print("Mode of traditional medicines:", traditional_mode)
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=col)
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```
